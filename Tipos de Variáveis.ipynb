# Tipos de Variavéis em Python

### Uma variável é um objecto, afinal tudo é um objeto em Python

          
- int -> inteiro
- string ou str -> texto
- float -> números om casas decimais (Ponto flutuante)
- bool ou beelean -> True ou False

Obs: Variáveis em Python não são declaradas explicitamente, mas isso não significa que você deve ficar mudando o tempo todo
